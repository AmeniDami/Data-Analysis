```python
#coubes multiples dans une meme figure

import matplotlib.pyplot as plt
import numpy as np


plt.subplots(figsize=(8,4))
x= np.linspace(0, 10)

plt.plot(x, np.sin(x), color='blue', linestyle='solid', label='sin(x)')

plt.plot(x, np.sin(x - 1), color='g', linestyle='dashed', label='sin(x - 1)')

plt.plot(x, np.sin(x - 2), color='0.75', linestyle='dashdot', label='sin(x - 2)')

plt.plot(x, np.sin(x - 3), color='#FF0000', linestyle='dotted', label='sin(x - 3)')
plt.axis('auto')

#Titre des axes
plt.xlabel('x' , color='r')
plt.ylabel('sin(x)' , color='r')

#title
plt.title("Intresting\nGraph")

#la legende dans laquel est affich les label qui sont dans plt.plot() 
plt.legend(loc='upper right')

#affichage du graphe
plt.show()

```

```python
#Exemple : ngative et positif 
#Remplissage du tableau par une distribution normale de valeurs
arr1 = np.random.randn(8,5)
#tab str de taille 7 caractres ayant le meme shape que arr1
arr2 = np.empty_like(arr1, dtype='U7') 
print("arr1 = ",arr1,"\n")
print("arr2 = ",arr2)
```

```python
#parcourir le tab ligne par ligne et colonne par colonne
for i in range(arr1.shape[0]): #returns the number of rows 
    for j in range(arr1.shape[1]): #returns the number of columns
        if arr1[i,j]<0:
            arr2[i,j] = "Negatif"
        else:
            arr2[i,j] = "Positif"
print(arr2)
        
```

Analyse descriptive avec numpy

```python
import numpy as np
x = np.arange(9)
arr = np.resize(x, (3,3))
arr
```

```python
print("Max: ", np.max(arr))
```

```python
#Maximum de chaque ligne
print("Max: ", np.max(arr,axis=0))
```

```python
#Maximum de chaque colonne
print("Max: ", np.max(arr,axis=1))
```

```python
#Minimum de chaque colonne
print("Min: ", np.min(arr,axis=1))
```

```python
#Somme de tous les lements
print("Sum: ", np.sum(arr))
```

```python
#Somme cumulatif
print("somme cumulative : \n",np.cumsum(arr))
print("somme cumulative de chaque colonne: \n",np.cumsum(arr, axis=0))
print("somme cumulative de chaque ligne: \n",np.cumsum(arr, axis=1))
```

```python
#moyenne arithmtique
print("la moyenne des lments : ", np.around(np.mean(arr),3))
```

```python
#moyenne par colonne
print("la moyenne des lments par colonne : ", np.around(np.mean(arr,axis=0),3))
```

```python
#moyenne par ligne
print("la moyenne des lments par ligne : ", np.around(np.mean(arr,axis=1),3))
```

```python
#la mdiane
print("mediane: ", np.median(arr))
#la mdiane par colonne
print("mediane par colonne: ", np.median(arr , axis=0))
#la mdiane par ligne
print("mediane par ligne: ", np.median(arr, axis=1))
```

```python
#l'cart-type 
print("l'cart-type: ", np.around(np.std(arr),3))
#l'cart-type par colonne
print("l'cart-type: ", np.around(np.std(arr , axis=0),3))
#l'cart-type par ligne
print("l'cart-type: ", np.around(np.std(arr, axis=1),3))
```

```python
#les variance
print("variance: ", np.around(np.var(arr),3))
#l'cart-type par colonne
print("variance: ", np.around(np.var(arr , axis=0),3))
#l'cart-type par ligne
print("variance: ", np.around(np.var(arr, axis=1),3))
```

```python
#Tableau avec des valeurs nulles 
import numpy as np
arr = np.array([[ 145, np.nan, 600],
               [ 400, 756, np.nan],
               [np.nan, 694, 236]])

print("Le plus grand lment dans le tableau: ",np.nanmax(arr))
print("Elments max par colonne: ", np.nanmax(arr, axis = 0))
print("Elments min par colonne: ", np.nanmin(arr, axis = 1))
print("Somme de tous les lments:", np.nansum(arr))

print("somme cumulative: \n", np.nancumsum(arr))
print("somme cumulative le long de chaque colonne: \n",np.nancumsum(arr, axis=0))
```
